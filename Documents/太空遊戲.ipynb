```python
import pygame
import random #"隨機"模組

FPS = 60
WIDTH = 500
HEIGHT = 600

#(Red,Green,Blue)調色盤，可填0-255，數字愈大顏色愈重
BlueGray = (119,136,153)
Black = (0,0,0)
RED = (255,0,0)
Yellow = (255,255,0)

#遊戲初始化 and 創建視窗
pygame.init()
screen = pygame.display.set_mode((500,600))
pygame.display.set_caption("太空遊戲")
clock = pygame.time.Clock()

#創建一個類別player
class Player(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self) #建平面
        self.image = pygame.Surface((50,40)) #長，寬
        self.image.fill(Black)
        self.rect = self.image.get_rect() #將平面框起來做定位
        #定位
        self.rect.centerx = WIDTH/2
        self.rect.y = 550
        #正中央:self.rect.center = (WIDTH/2, HEIGHT/2)
        self.speedx = 8
        
    #設定update函式 
    def update(self):
        key_pressed = pygame.key.get_pressed() #回傳布林值的函式(判斷是否有按鍵被按下，有則布林值為ture；無則為false)
        if key_pressed[pygame.K_RIGHT]:
            self.rect.x += self.speedx
        if key_pressed[pygame.K_LEFT]:
            self.rect.x -= self.speedx
        #防止方塊超出視窗
        if self.rect.right > WIDTH:
            self.rect.right = WIDTH
        if self.rect.left < 0:
            self.rect.left = 0
            
    def shoot(self):
        bullet = Bullet(self.rect.centerx, self.rect.top)
        all_sprites.add(bullet)

#創建一個類別Rock            
class Rock(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self) #建平面
        self.image = pygame.Surface((30,40)) #長，寬
        self.image.fill(RED)
        self.rect = self.image.get_rect() #將平面框起來做定位
        #定位
        self.rect.x = random.randrange(0, WIDTH - self.rect.width)
        self.rect.y = random.randrange(-100, -40)
        self.speedy = random.randrange(2, 10)
        self.speedx = random.randrange(-3, 3)
        
    #設定update函式 
    def update(self):
        self.rect.y += self.speedy
        self.rect.x += self.speedx
        #超出底部或左右邊界時，石頭重置
        if self.rect.top > HEIGHT or self.rect.left > WIDTH or self.rect.right < 0:
            self.rect.x = random.randrange(0, WIDTH - self.rect.width)
            self.rect.y = random.randrange(-100, -40)
            self.speedy = random.randrange(2, 10)
            self.speedx = random.randrange(-3, 3)

#創建一個類別Bullet
class Bullet(pygame.sprite.Sprite):
    def __init__(self, x, y):
        pygame.sprite.Sprite.__init__(self) #建平面
        self.image = pygame.Surface((10,20)) #長，寬
        self.image.fill(Yellow)
        self.rect = self.image.get_rect() #將平面框起來做定位
        #定位
        self.rect.centerx = x
        self.rect.bottom = y
        self.speedy = -10
        
    #設定update函式 
    def update(self):
        self.rect.y += self.speedy
        if self.rect.bottom < 0:
            self.kill()
            
#創建sprite群組
all_sprites = pygame.sprite.Group()

player = Player() #前面設定好的小物件player
all_sprites.add(player) #將小物件加到sprite群組裡
for i in range(8): #1次掉8顆石頭
    r = Rock()
    all_sprites.add(r)

#遊戲迴圈
running = True
while running:
    clock.tick(FPS) #1秒內只能被執行60(FPS)次
    #玩家輸入
    for event in pygame.event.get(): #回傳輸入者的動作(ex:按下空白建)
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                player.shoot()
        
    #更新遊戲
    all_sprites.update() #執行群組內所有物件的update函式
    
    #畫面顯示
    screen.fill(BlueGray) #背景顏色
    all_sprites.draw(screen) #將sprite群組中所有物件都畫在螢幕上
    pygame.display.update()
    
pygame.quit()
```
