```python
import pygame
import random
import time

# 初始化Pygame
pygame.init()

# 設定視窗大小
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("隨機射擊遊戲")

# 設定顏色
white = (255, 255, 255)
green = (0, 255, 0)
brown = (139, 69, 19)

# 設定松樹位置
tree_width = 60
tree_height = 100
tree_x = (screen_width - tree_width) // 2
tree_y = screen_height - tree_height

# 設定遊戲主迴圈
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # 隨機設定射擊火力
    shoot_power = random.randint(1, 10)

    # 設定松果的起始位置
    acorn_width = 20
    acorn_height = 20
    acorn_x = random.randint(0, screen_width - acorn_width)
    acorn_y = 0
    acorn_speed = 5

    # 開始松果掉落
    for _ in range(10):
        while acorn_y < screen_height:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    running = False
            
            acorn_y += acorn_speed

            # 清空畫面
            screen.fill(white)

            # 繪製松樹
            pygame.draw.rect(screen, brown, (tree_x, tree_y, tree_width, tree_height))
            pygame.draw.rect(screen, green, (tree_x - tree_width // 2, tree_y - 10, tree_width * 2, 10))

            # 繪製松果
            pygame.draw.rect(screen, brown, (acorn_x, acorn_y, acorn_width, acorn_height))

            # 顯示射擊火力
            font = pygame.font.Font(None, 36)
            text = font.render(f"射擊火力: {shoot_power}", True, (0, 0, 0))
            screen.blit(text, (10, 10))

            # 更新視窗
            pygame.display.flip()

            time.sleep(0.0001)  # 控制松果掉落速度，這裡是每0.1秒移動一次

    # 等待一段時間再重置遊戲
    time.sleep(2)

# 關閉Pygame
pygame.quit()
```

```python
import pygame
import random
import math

# 初始化Pygame
import pygame
import random
import math

# 初始化Pygame
pygame.init()

# 設定視窗大小
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("松果發射遊戲")

# 設定顏色
white = (255, 255, 255)
green = (0, 255, 0)
brown = (139, 69, 19)

# 設定松樹位置
tree_width = 60
tree_height = 100
tree_x = (screen_width - tree_width) // 2
tree_y = screen_height - tree_height

# 設定發射點位於視窗上方中點位置
shooter_x = screen_width // 2
shooter_y = 0

# 設定射擊速度
shooter_speed = 2

# 設定目標位置（可根據需要更改）
target_x = tree_x + tree_width // 2
target_y = tree_y

# 初始化射擊數據
shoot_data = []

# 計算射擊方向
shoot_angle = math.atan2(target_y - shooter_y, target_x - shooter_x)
shooter_dx = shooter_speed * math.cos(shoot_angle)
shooter_dy = shooter_speed * math.sin(shoot_angle)
acorn_x = shooter_x
acorn_y = shooter_y

# 設定遊戲主迴圈
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # 清空畫面
    screen.fill(white)

    # 繪製松樹
    pygame.draw.rect(screen, brown, (tree_x, tree_y, tree_width, tree_height))
    pygame.draw.rect(screen, green, (tree_x - tree_width // 2, tree_y - 10, tree_width * 2, 10))

    # 開始射擊
    while 0 <= acorn_x <= screen_width and 0 <= acorn_y <= screen_height:
        acorn_x += shooter_dx
        acorn_y += shooter_dy

        # 檢查是否擊中松樹
        hit_tree = acorn_x < tree_x + tree_width and acorn_x > tree_x and acorn_y > tree_y
        if hit_tree:
            print("擊中松樹！")
            running = False

        # 儲存射擊數據
        shoot_data.append({
            "Position": (int(acorn_x), int(acorn_y)),
            "Speed": shooter_speed,
            "Direction": math.degrees(shoot_angle),
            "Hit_Tree": hit_tree
        })

        # 繪製松果
        pygame.draw.rect(screen, brown, (int(acorn_x), int(acorn_y), 10, 10))

        # 更新視窗
        pygame.display.flip()

# 顯示射擊數據
for i, data in enumerate(shoot_data):
    print(f"Step {i+1}:")
    print(f"  Position: {data['Position']}")
    print(f"  Speed: {data['Speed']}")
    print(f"  Direction: {data['Direction']} degrees")
    print(f"  Hit Tree: {data['Hit_Tree']}")
    print()

# 關閉Pygame
pygame.quit()
```
