```python
import pygame
import random

FPS = 60
WIDTH = 800
HEIGHT = 500

#(Red,Green,Blue)調色盤，可填0-255，數字愈大顏色愈重
BlueGray = (119,136,153)
Black = (0,0,0)
Red = (255,0,0)
Yellow = (255,255,0)

#遊戲初始化 and 創建視窗
pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("跑酷遊戲")
clock = pygame.time.Clock() #遊戲速度

class Player(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self) #建平面
        self.image = pygame.Surface((50,40)) #長，寬
        self.image.fill(Black)
        self.rect = self.image.get_rect() #將平面框起來做定位
        #定位
        self.rect.x = 50
        self.rect.y = 450
        
   #設定update函式 
    def jump(self):
        key_pressed = pygame.key.get_pressed() #回傳布林值的函式(判斷是否有按鍵被按下，有則布林值為ture；無則為false)
        jump_height = 150  # 跳躍高度
        is_jumping = False #追蹤玩家是否在跳躍中
        current_time = pygame.time.get_ticks()  # 取得當前時間的毫秒數
        jump_start_time = 0 # 記錄跳躍開始的時間
        # 檢測按鍵事件，以實現跳躍
        if key_pressed[pygame.K_SPACE] and not is_jumping:
            is_jumping = True
            jump_start_time = current_time  # 記錄跳躍開始的時間
        # 如果正在跳躍，更新角色的位置
        if is_jumping:
            self.rect.y -= jump_height
        # 如果跳躍時間超過5秒，恢復到原位
        if current_time - jump_start_time >= 5000:
            self.rect.y = self.original_y # 回到原位
            is_jumping = False 
            
class Obstacle(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self) #建平面
        self.image = pygame.Surface((50,300)) #長，寬
        self.image.fill(Red)
        self.rect = self.image.get_rect() #將平面框起來做定位
        #定位
        self.rect.x = random.randrange(800, 820)
        self.rect.y = 400     
        self.speedx = 4
    
    def update(self):
        self.rect.x -= self.speedx
        if self.rect.left < -60:
            self.rect.x = random.randrange(800, 820)
            self.rect.y = 400     
            self.speedx = 4
        
#創建sprite群組
all_sprites = pygame.sprite.Group()
player = Player() #前面設定好的小物件player
all_sprites.add(player) #將小物件加到sprite群組裡
for i in range(5):
    obstacle = Obstacle()
    all_sprites.add(obstacle)

#遊戲迴圈
running = True
while running:
    clock.tick(FPS) #1秒內迴圈只能被執行60(FPS)次
    #取得輸入
    for event in pygame.event.get(): #回傳輸入者的動作(ex:按下空白建)
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                player.jump()
        
    #更新遊戲
    all_sprites.update() #執行群組內所有物件的update函式
    
    #畫面顯示
    screen.fill(BlueGray) #背景顏色
    all_sprites.draw(screen) #將sprite群組中所有物件都畫在螢幕上
    pygame.display.update() #畫面更新
    
pygame.quit()
```
